## Part 1 - XOR
1. Using the XOR dataset below, train (400 epochs) a neural network (NN) using 1, 2, 3, 4, and 5 hidden layers (where each layer has only 2 neurons). For each n layers, store the resulting loss score along with n. Plot the results to find what the optimal number of layers is.

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import SGD  #Stochastic Gradient Descent

import numpy as np
# fix random seed for reproducibility
np.random.seed(7)

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 40
xx = np.random.random((n,1))
yy = np.random.random((n,1))
```

```python
X = np.array([np.array([xx,-xx,-xx,xx]),np.array([yy,-yy,yy,-yy])]).reshape(2,4*n).T
y = np.array([np.ones([2*n]),np.zeros([2*n])]).reshape(4*n)
```

```python
def run_XOR_nn(num_layers, n_neurons_per_hlayer, act_function, optimizer):
    # num_layers = [1,2,3,4,5]
    scores = []
    for num_layer in num_layers:

        # build model and evaluate
        model = Sequential()
        for i in range(num_layer):
            if i == 0:
                model.add(Dense(n_neurons_per_hlayer, input_dim=2, activation=act_function))  #sigmoid, relu
            else:
                model.add(Dense(n_neurons_per_hlayer, activation=act_function))
        model.add(Dense(1, activation='sigmoid'))
        if optimizer == 'sgd':
            sgd = SGD(learning_rate=0.1)
        else:
            pass
        model.compile(loss='binary_crossentropy', optimizer=optimizer)

        model.fit(X, y, batch_size=2, epochs=400)

        # evaluate the model
        score = model.evaluate(X, y)

        scores.append(score)
        
    return(num_layers, scores)
```

```python
num_layers, scores = run_XOR_nn(num_layers= [1,2,3,4,5],
                               n_neurons_per_hlayer=2,
                               act_function='tanh',
                               optimizer='sgd')
```

```python
# plot scores
plt.scatter(num_layers, scores)
plt.xticks(num_layers)
plt.xlabel('Number of Hidden Layers')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 2 Neurons per Hidden Layer')
```

The optimal number of hidden layers in this configuration is 5.

2. Repeat the above with 3 neurons in each Hidden layers. How do these results compare to the 2 neuron layers?

```python
num_layers, scores = run_XOR_nn(num_layers= [1,2,3,4,5],
                               n_neurons_per_hlayer=3,
                               act_function='tanh',
                               optimizer='sgd')
```

```python
# plot scores
plt.scatter(num_layers, scores)
plt.xticks(num_layers)
plt.xlabel('Number of Hidden Layers')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 3 Neurons per Hidden Layer')
```

Unlike the NN with 2 Neurons per Hidden Layer, the NN with 3 Neurons per Hidden Layer has a minimum loss at n = 4 Hidden Layers (5 Hidden Layers had the minimum loss for the NN with 2 Neurons per Hidden Layer).

3. Repeat the above with 4 neurons in each Hidden layers. How do these results compare to the 2 and 3 neuron layers?

```python
num_layers, scores = run_XOR_nn(num_layers= [1,2,3,4,5],
                               n_neurons_per_hlayer=4,
                               act_function='tanh',
                               optimizer='sgd')
```

```python
# plot scores
plt.scatter(num_layers, scores)
plt.xticks(num_layers)
plt.xlabel('Number of Hidden Layers')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 4 Neurons per Hidden Layer')
```

The minimum loss for a NN with 4 Neurons per Hidden Layer is found at n = 3 hidden layers.  The loss here is less than the minimum loss for NNs with 2 and 3 Neurons per Hidden Layer.  Therefore, the "most optimal" configuration for question 4 will comprise of a NN with 3 Hidden Layers and 4 Neurons per Hidden Layer.

4. Using the most optimal configuraion (n-layers, k-neurons per layer), compare how `tanh`, `sigmoid`,`softplus` and `relu` effect the loss after 400 epochs. Try other Activation functions as well (https://keras.io/activations/)

```python
act_functions_to_test = ['tanh', 'sigmoid', 'softplus', 'relu', 'selu', 'elu']
scores_agg = []
for act_func in act_functions_to_test:
    num_layers, scores = run_XOR_nn(num_layers= [3],
                                   n_neurons_per_hlayer=4,
                                   act_function=act_func,
                                   optimizer='sgd')
    for s in scores:
        scores_agg.append(s)
```

```python
len(act_functions_to_test), len(scores_agg)
```

```python
# plot scores
plt.scatter(act_functions_to_test, scores_agg)
plt.xlabel('Activation Function')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 3 Hidden Layers and 4 Neurons per Hidden Layer')
```

```python
list(zip(act_functions_to_test, scores_agg))
```

The Activation Function with the minimal loss was tanh with a 0.0313 loss (binary crossentropy).

5. Again with the most optimal setup, try other optimizers (instead of `SGD`) and report on the loss score. (https://keras.io/optimizers/)

```python
optimizers_to_test = ['sgd', 'Adam', 'RMSprop', 'Ftrl', 'Nadam']
scores_agg = []
for optim in optimizers_to_test:
    num_layers, scores = run_XOR_nn(num_layers= [3],
                                   n_neurons_per_hlayer=4,
                                   act_function='tanh',
                                   optimizer=optim)
    for s in scores:
        scores_agg.append(s)
```

```python
len(optimizers_to_test), len(scores_agg)
```

```python
# plot scores
plt.scatter(optimizers_to_test, scores_agg)
plt.xlabel('Optimizer')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 3 Hidden Layers and 4 Neurons per Hidden Layer (tanh activation function)')
```

```python
list(zip(optimizers_to_test, scores_agg))
```

The optimizer 'Adam' has the minimum loss, 0.01986.

In summary, a NN with 3 Hidden Layers, 4 Neurons per Hidden Layer, a tanh activation function in each hidden layer, and an optimizer of type 'Adam' has the minimum loss of all models tested.

### Predict Presidential Preference from race, income, sex, and urban

I will be using the 2020 ANES study; specifically, the cleaned dataset that I created in the 'Unleashing Open Data with Python' course last semester.  In order to minimize the amount of code in this jupyter notebook, I left all of the cleaning/recoding code in the previous jupyter notebook, exported the cleaned dataframe to a .csv, and will import the cleaned .csv below.

```python
import pandas as pd
```

```python
# read in dataset
ANES = pd.read_csv(r'C:\Users\17036\Desktop\mlnn-personal\Week-9\ANES_2020.csv', index_col=0)
ANES.head()
```

I'm only taking a sample of 1500 rows from this dataset to speed up model run time.  1000 observations will be used for model training and 500 for testing.

```python
# transform columns
from sklearn import preprocessing, decomposition
from sklearn.model_selection import train_test_split
ANES_clean = ANES.copy()
ANES_clean = ANES_clean.drop('social_class', axis=1)
non_numerical_cols = ['race', 'sex', 'urban']

ord_enc = preprocessing.OrdinalEncoder()
ANES_clean[non_numerical_cols] = ord_enc.fit_transform(ANES_clean[non_numerical_cols])
ANES_clean = ANES_clean.sample(n = 1500)
print(ANES_clean)
```

```python
X_ANES = np.array(ANES_clean.drop('president', axis=1))
y_ANES = np.array(ANES_clean['president'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_ANES, y_ANES, test_size=0.33)
```

```python
def run_ANES_nn(num_layers, n_neurons_per_hlayer, act_function, optimizer, x_data, y_data):
    # num_layers = [1,2,3,4,5]
    scores = []
    for num_layer in num_layers:

        # build model and evaluate
        model = Sequential()
        for i in range(num_layer):
            if i == 0:
                model.add(Dense(n_neurons_per_hlayer, input_dim=4, activation=act_function))  #sigmoid, relu
            else:
                model.add(Dense(n_neurons_per_hlayer, activation=act_function))
        model.add(Dense(1, activation='sigmoid'))
        if optimizer == 'sgd':
            sgd = SGD(learning_rate=0.1)
        else:
            pass
        model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

        model.fit(x_data, y_data, batch_size=10, epochs=400)

        # evaluate the model
        score = model.evaluate(x_data, y_data)

        scores.append(score)
        
    return(num_layers, scores)
```

```python
num_layers, scores = run_ANES_nn(num_layers= [1,2,3],
                                 n_neurons_per_hlayer=2,
                                 act_function='tanh',
                                 optimizer='sgd',
                                 x_data=X_train,
                                 y_data=y_train)
```

```python
# plot scores
plt.scatter(num_layers, [i[0] for i in scores])
plt.xticks(num_layers)
plt.xlabel('Number of Hidden Layers')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 2 Neurons per Hidden Layer')
```

3 Hidden Layers appears to have the minimum loss; however, the loss for each of the 3 models is very similar.

```python
num_layers, scores = run_ANES_nn(num_layers= [1,2,3],
                                 n_neurons_per_hlayer=4,
                                 act_function='tanh',
                                 optimizer='sgd',
                                 x_data=X_train,
                                 y_data=y_train)
```

```python
# plot scores
plt.scatter(num_layers, [i[0] for i in scores])
plt.xticks(num_layers)
plt.xlabel('Number of Hidden Layers')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 4 Neurons per Hidden Layer')
```

```python
[i[0] for i in scores]
```

Increasing the number of Neurons per Hidden Layer to 4 doesn't appear to reduce the loss any further.

```python
num_layers, scores = run_ANES_nn(num_layers= [1,2,3],
                                 n_neurons_per_hlayer=6,
                                 act_function='tanh',
                                 optimizer='sgd',
                                 x_data=X_train,
                                 y_data=y_train)
```

```python
# plot scores
plt.scatter(num_layers, [i[0] for i in scores])
plt.xticks(num_layers)
plt.xlabel('Number of Hidden Layers')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 6 Neurons per Hidden Layer')
```

3 Hidden Layers appears to have the minimum loss.  I will continue forward with 6 Neurons per Hidden Layer and 3 Hidden Layers.

```python
act_functions_to_test = ['tanh', 'relu', 'LeakyReLU', 'elu']
scores_agg = []
for act_func in act_functions_to_test:
    num_layers, scores = run_ANES_nn(num_layers= [3],
                                   n_neurons_per_hlayer=6,
                                   act_function=act_func,
                                   optimizer='sgd',
                                   x_data=X_train,
                                   y_data=y_train)
    for s in scores:
        scores_agg.append(s)
```

```python
# plot scores
plt.scatter(act_functions_to_test, [i[0] for i in scores_agg])
plt.xlabel('Activation Function')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 3 Hidden Layers and 6 Neurons per Hidden Layer')
```

Relu appears to have the minimum loss.  LeakyReLU did not calculate (likely need to wrap it in the Activation function first).

```python
optimizers_to_test = ['sgd', 'Adam', 'Nadam']
scores_agg = []
for optim in optimizers_to_test:
    num_layers, scores = run_ANES_nn(num_layers= [3],
                                   n_neurons_per_hlayer=6,
                                   act_function='tanh',
                                   optimizer=optim,
                                   x_data=X_train,
                                   y_data=y_train)
    for s in scores:
        scores_agg.append(s)
```

```python
# plot scores
plt.scatter(optimizers_to_test, [i[0] for i in scores_agg])
plt.xlabel('Optimizer')
plt.ylabel('Loss (binary crossentropy)')
plt.title('NN with 3 Hidden Layers and 6 Neurons per Hidden Layer (tanh activation function)')
```

Adam appears to have the minimum loss of the tested optimizers.

In summary, of the values tested, a NN with 3 Hidden Layers, 6 Neurons per Hidden Layer, the Relu activation function, and the Adam optimizer produce a model with the lowest loss.  It is unlikely these are the best values, but due to the size of the dataset and time constraints, the batch size was set relatively high (10) and only 400 epochs were used.  In order to further optimize this NN, it would be best to test models with more neurons, more hidden layers, and smaller batch sizes (also possibly more epochs).

In order to see how the NN performs on the test set:

```python
model = Sequential()
model.add(Dense(6, input_dim=4, activation='relu'))  #sigmoid, relu
model.add(Dense(6, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])

model.fit(X_train, y_train, batch_size=10, epochs=400)
```

```python
scores = model.evaluate(X_train, y_train)
print('Train set:')
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

scores = model.evaluate(X_test, y_test)
print('Test set:')
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

This NN performs slightly better on the test set than the train set.
