# Neural Networks - Image Recognition 

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
from keras.datasets import mnist
from keras.models import Sequential
from tensorflow.keras.optimizers import RMSprop
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend
```

```python
import matplotlib.pyplot as  plt
%matplotlib inline
```

## Conv Net
Trains a simple convnet on the MNIST dataset.
Gets to 99.25% test accuracy after 12 epochs
(there is still a lot of margin for parameter tuning).


```python
# input image dimensions
img_rows, img_cols = 28, 28

# the data, shuffled and split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

if backend.image_data_format() == 'channels_first':
    x_train = x_train.reshape(x_train.shape[0], 1, img_rows, img_cols)
    x_test = x_test.reshape(x_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
    x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
scales_to_test = ['none', .1, .5, 1.0, 2.0, 4.0]
train_accuracies = []
test_accuracies = []

batch_size = 128
num_classes = 10
epochs = 12

# convert class vectors to binary class matrices
y_train = tf.keras.utils.to_categorical(y_train, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)

for scale in scales_to_test:
    # add some random noise to both x_train and x_test
    if scale == 'none':
        x_train_new = x_train
        x_test_new = x_test
    else:
        x_train_new = x_train + np.random.normal(scale = scale, size=x_train.shape)
        x_test_new = x_test + np.random.normal(scale = scale, size=x_test.shape)

        
    model = Sequential()
    model.add(Conv2D(32, kernel_size=(3, 3),
                     activation='relu',
                     input_shape=input_shape))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))

    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=tf.keras.optimizers.Adadelta(learning_rate=0.01),
                  metrics=['accuracy'])

    history = model.fit(x_train_new, y_train,
                      batch_size=batch_size,
                      epochs=epochs,
                      verbose=1,
                      validation_data=(x_test_new, y_test))
    score = model.evaluate(x_test_new, y_test, verbose=0)

    train_accuracies.append(history.history['accuracy'][-1])
    test_accuracies.append(score[1])  
```

```python
print('The train accuracy without noise is: ' + str(round(train_accuracies[0], 4)))
print('The train accuracy with noise (scale = 0.1) is: ' + str(round(train_accuracies[1], 4)))
print('The test accuracy without noise is: ' + str(round(test_accuracies[0], 4)))
print('The test accuracy with noise (scale = 0.1) is: ' + str(round(test_accuracies[1], 4)))
```

```python
# graph of Train and Test Accuracies for each scale (of noise)
plt.figure()
plt.plot(scales_to_test, train_accuracies, label = 'Train Data')
plt.plot(scales_to_test, test_accuracies, label = 'Test Data')
plt.xlabel('Scale (standard deviation of noise)')
plt.ylabel('Model Accuracy')
plt.title('NN Model Accuracy on Image Classification with added noise')
plt.grid(True)
plt.legend()
plt.show()
```

Comparing the results of the MultiLayer Perceptron to the ConvNet can be seen below (for each level of noise).  It should be noted the training time for the ConvNet was significantly more than for the MultiLayer Perceptron.

```python
train_accuracies_perc = [0.9949, 0.9979, 0.9944, 0.9769, 0.8920, 0.6679]
test_accuracies_perc = [0.9835, 0.9758, 0.9113, 0.7559, 0.4059, 0.1898]
```

```python
# graph of Train and Test Accuracies for each scale (of noise)
plt.figure()
plt.plot(scales_to_test, train_accuracies, label = 'Train Data (ConvNet)')
plt.plot(scales_to_test, test_accuracies, label = 'Test Data (ConvNet)')
plt.plot(scales_to_test, train_accuracies_perc, label = 'Train Data (Perceptron)')
plt.plot(scales_to_test, test_accuracies_perc, label = 'Test Data (Perceptron)')
plt.xlabel('Scale (standard deviation of noise)')
plt.ylabel('Model Accuracy')
plt.title('NN Model Accuracy on Image Classification with added noise')
plt.grid(True)
plt.legend()
plt.show()
```

The accuracy for the Perceptron model was higher when using the training set; when using the test set, the Perceptron NN outperformed ConvNet for low values of noise (scale <= 1), but was surpassed for higher values of noise (scale > 1).
